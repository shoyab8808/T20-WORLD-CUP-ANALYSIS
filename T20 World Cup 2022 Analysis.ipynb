```python
#first inning wicket
plt.figure(figsize=(12,6))
sns.countplot(x="first innings wickets", data = df, hue = "toss winner")
plt.title("first innings wickets", fontsize=10)
plt.xlabel("MATCH", fontsize=10)
plt.ylabel("WICKET", fontsize=10)
plt.show()
```

```python
#number of matches winn by team
plt.figure(figsize=(12, 6))
sns.countplot(x="winner", data = df, palette="cool")
plt.title("MATCH WINNER", fontsize=10)
plt.xticks(rotation=75)
plt.xlabel("TEAM")
plt.ylabel("NUMBER OF WINNS")
plt.show()
```

```python
WIN = df.winner.value_counts()
LABELS = np.array(WIN.index)
SIZES = WIN.values
COLORS = ["#FFBF00", "#FA8072"]
plt.figure(figsize=(10, 8))
plt.pie(SIZES, labels = LABELS, colors = COLORS, autopct="%1.1f%%", shadow = True, startangle=90)
plt.title("Match Result", fontsize=20)
plt.axis("equal")
plt.show
```

```python
TOSS = df["toss decision"].value_counts()
LABELS = np.array(TOSS.index)
SIZES = TOSS.values
COLORS=["#FFBF00", "#FA8072"]
plt.figure(figsize=(10, 8))
plt.pie(SIZES, labels = LABELS, colors=COLORS, autopct = "%1.1f%%", shadow=True, startangle = 90)
plt.title("Toss Result")
plt.axis("equal")
plt.show()
```

```python
df.columns.values
```

```python
# we will print winner season wise
final_matches = df.drop_duplicates(subset=["winner"], keep = "last")
final_matches[["venue", "winner"]].reset_index(drop=True).sort_values("winner")
```

```python
#we will plot pie chart on winning percentage in final
MATCH = final_matches["won by"].value_counts()
LABELS = np.array(TOSS.index)
SIZES = MATCH.values 
COLORS = ["gold", "purple"]
plt.figure(figsize=(10, 8))
plt.pie(SIZES, labels = LABELS, colors = COLORS, autopct = "%1.1f%%", shadow = True, startangle=90)
plt.title("Match Result")
plt.axis("equal")
plt.show()
```

```python
#we will print name of top player in t20 world cup
plt.figure(figsize = (18, 10))
top_players = df["player of the match"].value_counts()[:10]
fig, ax = plt.subplots()
ax.set_ylim([0,10])
ax.set_ylabel("count")
ax.set_title("Top player of the match Winners")
top_players.plot.bar()
sns.barplot(x = top_players.index, y = top_players, orient = "v", palette = "hsv");
plt.show();
```
